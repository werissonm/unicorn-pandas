```python
import pandas as pd
data = pd.read_csv("C:\\Users\\werim\\OneDrive\\Documentos\\MasterSchool\\project_da_105\\unicorn_data.csv")
data.head()
 
```

```python

# Use regular expressions to extract the desired components
data[['Country', 'Year', 'Item ID', 'Customer ID']] = data['Order ID'].str.extract(r'([A-Z]{2})-(\d{4})-(\d{3})(\d{3})')

# Convert 'Year' column to datetime format for just the year
data['Year'] = pd.to_datetime(data['Year'], format='%Y').dt.year
data
```

```python
# total amount of customers
total_customers = data['Customer Name'].nunique()
total_customers
```

```python
total_cities = data['City'].nunique()
total_cities
```

```python
# Find the city with the highest sales
top_sales_city = data.groupby('City')['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
top_sales_city
```

```python
# Calculate the average discount
average_discount = (data['Discount'].mean()*100).round(1)
average_discount
```

```python
# Filter the DataFrame for the "Consumer" segment
consumer_df = data[data['Segment'] == 'Consumer']

# Group by product and calculate the count
product_counts = consumer_df['Product Name'].value_counts().reset_index()

# Rename the columns for clarity
product_counts.columns = ['Product Name', 'Count']
product_counts
```

```python
# Filter the DataFrame for the "Office Supplies" category
office_supplies_df = data[data['Category'] == 'Office Supplies']

# Calculate the total profit
total_profit = office_supplies_df['Profit'].sum()
total_profit
```

```python
# Group by customer name and count orders
customer_order_counts = data.groupby('Customer Name')['Order ID'].count()

# Find the customer(s) with the most orders
customers_with_most_orders = customer_order_counts[customer_order_counts == customer_order_counts.max()]
customers_with_most_orders
```

```python
# Group by state and sum profits
state_profit = data.groupby('State')['Profit'].sum()
pd.DataFrame(state_profit)

# Find the state with the highest total profit
state_with_highest_profit = state_profit.idxmax()
state_with_highest_profit
```

```python
# Filter orders with 'Standard Class' ship mode
standard_class_orders = data[data['Ship Mode'] == 'Standard Class']
standard_class_orders

# Count the number of orders in 'Standard Class'
num_standard_class_orders = len(standard_class_orders)
num_standard_class_orders
```

```python
# Convert the 'Unit Price' column to float
data['unit price'] = pd.to_numeric(data['unit price'], errors='coerce')

# Calculate the price per unit before discounts
data['Price per Unit (Before Discount)'] = (data['unit price'] / (1 - data['Discount']))
data
```

```python
# Find the product with the highest unit price
most_expensive_product = data[data['Product Name'] == data['Price per Unit (Before Discount)'].max()]
most_expensive_product
```

```python
# Create a pivot table to calculate total sales for each manufacturer and category combination
pivot_table = pd.pivot_table(data, values='Sales', index=['Category', 'Manufacturer'], aggfunc='sum').reset_index()
pivot_table

# Filter the pivot table for the 'Technology' category
technology_sales = pivot_table[pivot_table['Category'] == 'Technology']
technology_sales

# Sort the 'Total Sales' column in descending order
sorted_technology_sales = technology_sales.sort_values(by='Sales', ascending=False)

# Find the manufacturer with the second-highest sales
second_highest_manufacturer = sorted_technology_sales.iloc[1]['Manufacturer']
second_highest_manufacturer
```

```python
# Calculate profit margin and create a new column
data['Profit Margin'] = (data['Profit'] / data['Sales'] * 100).round(1)

# Calculate the average profit margin
average_profit_margin = data['Profit Margin'].mean()
average_profit_margin
```

```python
filtered_order = data[data['Order ID'] == 'CA-2015-100090']
filtered_order

correct_result = data['Customer_Location'] = data['Customer Name'] + ', ' + data['City'] + ', ' + data['State']
correct_result
```

```python
import numpy as np

# Define conditions
conditions = [
    (data['Sales'] > 200) & (data['Profit'] > 20),
    (data['Profit'] > 40),
    (data['Profit'] <= 0)
]

# Corresponding categories
categories = ['High', 'Low', 'Loss']

# Column based on conditions
data['Profit Category'] = np.select(conditions, categories, default='Low')

loss_count = len(data[data['Profit Category'] == 'Loss'])
loss_count


```
